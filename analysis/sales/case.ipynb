```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
%matplotlib inline

#Посчитаем суммарную выручку для каждого сотрудника
```

```python
df = pd.read_csv('sales.csv')
```

```python
df.head(20)
```

```python
len(df['employee_id'].unique())
```

```python
df['revenue'].groupby(df['employee_id']).sum()
```

```python
df['convers'] = (df['sales_count'] / df['leads_count']) * 100
df['convers'] = df['convers'].round()
df.head(10)
```

```python
conversion_count = df.groupby('employee_id')['convers'].sum()
```

```python

conversion_count.plot(kind='bar', figsize=(10, 6))

plt.title('Количество конверсий на сотрудника')
plt.xlabel('ID сотрудника')
plt.ylabel('Количество конверсий')

plt.show()
```
